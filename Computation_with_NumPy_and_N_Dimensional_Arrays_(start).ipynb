```python
import numpy as np
import matplotlib.pyplot as plt

# Create an image with different color channels
height, width = 100, 100
image_array = np.zeros((height, width, 3), dtype=np.uint8)

# Set the red channel to 255 in the top half
image_array[:height//2, :, 0] = 255  # Red

# Set the green channel to 255 in the bottom left quarter
image_array[height//2:, :width//2, 1] = 255  # Green

# Set the blue channel to 255 in the bottom right quarter
image_array[height//2:, width//2:, 2] = 255  # Blue

# Convert the RGB image to grayscale using the formula Y_linear
grayscale_image = (0.2126 * image_array[:, :, 0] + 
                   0.7152 * image_array[:, :, 1] + 
                   0.0722 * image_array[:, :, 2]).astype(np.uint8)

# Rotate the grayscale image 90 degrees counterclockwise
rotated_grayscale_image = np.rot90(grayscale_image)

# Display the rotated grayscale image
plt.imshow(rotated_grayscale_image, cmap='gray')
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an image with different color channels
height, width = 100, 100
image_array = np.zeros((height, width, 3), dtype=np.uint8)

# Set the red channel to 255 in the top half
image_array[:height//2, :, 0] = 255  # Red

# Set the green channel to 255 in the bottom left quarter
image_array[height//2:, :width//2, 1] = 255  # Green

# Set the blue channel to 255 in the bottom right quarter
image_array[height//2:, width//2:, 2] = 255  # Blue

# Invert the colors of the RGB image
inverted_image_array = 255 - image_array

# Display the inverted image
plt.imshow(inverted_image_array)
plt.axis('off')  # Turn off axis labels
plt.show()
```
